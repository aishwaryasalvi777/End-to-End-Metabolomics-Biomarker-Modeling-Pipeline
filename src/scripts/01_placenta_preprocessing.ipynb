```python
import sys
from pathlib import Path
import os

# Add project root to path
sys.path.insert(0, str(Path(os.getcwd()).parent.parent))

import pandas as pd
from src.utils.config import PLACENTA_ANNO, PLACENTA_DATA, PLACENTA_DATA_T
from src.utils.io_utils import load_placenta_normalized, save_csv
```

# Load Data

```python
print('-' * 80)
print(" Task 1: Loading placenta Data")
print('-' * 80)
df = load_placenta_normalized()
print(f"   Loaded shape: {df.shape}")
```

```python
df.head()
```

# Split Annotation vs Data

```python
print('-' * 80)
print(" Task 2: Splitting annotation vs metabolite data")
print('-' * 80)
df_anno = df.iloc[:, :11].copy()     # Columns A-K (0-10)
df_data = df.iloc[:, 11:].copy()    # Columns L-AY (11-end)

# Save both
save_csv(df_anno, PLACENTA_ANNO)
save_csv(df_data, PLACENTA_DATA)

print(f"   Annotation: {df_anno.shape}")
print(f"   Metabolite data: {df_data.shape}")
      
print(" Task 2 complete!")

```

# Preview before transpose

```python
print("Data shape before transpose:", df_data.shape)
print("Sample names (columns):", df_data.columns[:5].tolist())
print("Metabolite COMP IDs:", df_anno.iloc[:5, 0].tolist())
```

# Transpose & Save

```python
print('-' * 80)
print("🔄 Task 3: Transposing - Samples as ROWS, COMP ID as COLUMNS")
print('-' * 80)

# Get COMP ID (1st column) as strings
comp_ids = df_anno.iloc[:, 0].astype(str).tolist() 

# Transpose + set column names
df_data_t = df_data.T.reset_index()
df_data_t.columns = ['Sample'] + comp_ids

save_csv(df_data_t, PLACENTA_DATA_T)
# print(f"   Transposed: {PLACENTA_DATA_T} (shape: {df_data_t.shape})")
print(f"   First columns: ['Sample'] + {comp_ids[:4]}")
print(df_data_t.head())
```
