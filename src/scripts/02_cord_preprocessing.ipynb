# Imports & Setup

```python
import sys
from pathlib import Path
import os

# Add project root to path
sys.path.insert(0, str(Path(os.getcwd()).parent.parent))

import pandas as pd
from src.utils.config import CORD_ANNO, CORD_DATA, CORD_DATA_T
from src.utils.io_utils import load_cord_normalized, save_csv
```

# Load Data

```python
print('-' * 80)
print("🔄 Task 1: Loading cord blood data...")
print('-' * 80)
df = load_cord_normalized()
print(f"   Loaded shape: {df.shape}")

```

```python
print("Chord Blood Data: ")
df.head()
```

# Split Annotation vs Data

```python
print('-' * 80)
print("🔄 Task 2: Splitting Annotation vs Metabolite Data")
print('-' * 80)
df_anno = df.iloc[:, :11].copy()
df_data = df.iloc[:, 11:].copy()

# Save both
save_csv(df_anno, CORD_ANNO)
save_csv(df_data, CORD_DATA)

print(f"   Annotation saved: {CORD_ANNO} (shape: {df_anno.shape})")
print(f"   Metabolite data saved: {CORD_DATA} (shape: {df_data.shape})")
print(" Task 2 complete!")
```

# Preview Before Transpose

```python
print("Data shape before transpose:", df_data.shape)
print("Sample names (columns):", df_data.columns[:5].tolist())
print("Metabolite COMP IDs:", df_anno.iloc[:5, 0].tolist())
```

# Transpose & Save

```python
print('-' * 80)
print("🔄 Task 3: Transposing - Samples as ROWS, COMP ID as COLUMNS")
print('-' * 80)

# Get COMP ID as strings
comp_ids = df_anno.iloc[:, 0].astype(str).tolist() 

# Transpose + set column names
df_data_t = df_data.T.reset_index()
df_data_t.columns = ['Sample'] + comp_ids

save_csv(df_data_t, CORD_DATA_T)
print(f"   Transposed: {CORD_DATA_T} (shape: {df_data_t.shape})")
print(f"   First columns: ['Sample'] + {comp_ids[:4]}")
print(df_data_t.head())
```
